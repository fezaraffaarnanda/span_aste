```python
# SPAN-ASTE Complete Prediction Notebook
# This notebook includes the full model architecture for predictions

import os
import torch
import pandas as pd
import numpy as np
import itertools
from enum import IntEnum
from torch import nn
from transformers import BertTokenizer, BertModel, AlbertModel, AutoTokenizer

# Define enums needed for the model
class SpanLabel(IntEnum):
    INVALID = 0
    ASPECT = 1
    OPINION = 2

class RelationLabel(IntEnum):
    INVALID = 0
    POS = 1
    NEG = 2
    NEU = 3

# Configurations - Change these to match your setup
MODEL_PATH = "checkpoint/fold5/model_best"  # Path to model directory
BERT_MODEL = "indobenchmark/indobert-base-p2"  # Model used for training
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# Function to clean aspect terms
def clean_aspect_term(aspect):
    """
    Post-process aspect terms to remove common modifiers and function words
    that shouldn't be part of the aspect
    """
    # Common function words in Indonesian that often incorrectly follow aspect terms
    stop_words = ['nya', 'yg', 'yang', 'sih', 'kan', 'lah', 'dong', 'deh', 'kok', 'juga']
    
    # Split the aspect
    words = aspect.split()
    
    # If aspect is just one word, return it as is
    if len(words) <= 1:
        return aspect
    
    # Check if the last word is a common function word
    if words[-1].lower() in stop_words:
        return ' '.join(words[:-1])
    
    # For cases like "aplikasi rusak" where "rusak" is actually an opinion
    # Return just the first word if it's a common noun and second word looks like an adjective
    common_nouns = ['aplikasi', 'sistem', 'fitur', 'tampilan', 'menu', 'layanan', 'website', 'app']
    if words[0].lower() in common_nouns and len(words) == 2:
        return words[0]
    
    return aspect

# Function to clean opinion terms
def clean_opinion_term(opinion):
    """Clean opinion terms by removing special tokens and extra punctuation"""
    # Remove special tokens
    opinion = opinion.replace('[CLS]', '').replace('[SEP]', '').strip()
    
    # Remove trailing punctuation while preserving punctuation inside
    while opinion and opinion[-1] in [',', '.', '!', '?', ';', ':']:
        opinion = opinion[:-1].strip()
        
    return opinion

# Define model components to match the architecture of your trained model
class SpanRepresentation(nn.Module):
    def __init__(self, span_width_embedding_dim, span_maximum_length):
        super(SpanRepresentation, self).__init__()
        self.span_maximum_length = span_maximum_length
        self.bucket_bins = [0, 1, 2, 3, 4, 5, 7, 8, 15, 16, 31, 32, 63, 64]
        self.span_width_embedding = nn.Embedding(len(self.bucket_bins), span_width_embedding_dim)

    def bucket_embedding(self, width, device):
        em = [ix for ix, v in enumerate(self.bucket_bins) if width >= v][-1]
        return self.span_width_embedding(torch.LongTensor([em]).to(device))

    def forward(self, x, batch_max_seq_len):
        batch_size, sequence_length, _ = x.size()
        device = x.device

        len_arrange = torch.arange(0, batch_max_seq_len, device=device)
        span_indices = []

        max_window = min(batch_max_seq_len, self.span_maximum_length)

        for window in range(1, max_window + 1):
            if window == 1:
                indics = [(x.item(), x.item()) for x in len_arrange]
            else:
                res = len_arrange.unfold(0, window, 1)
                indics = [(idx[0].item(), idx[-1].item()) for idx in res]
            span_indices.extend(indics)

        spans = [torch.cat(
            (x[:, s[0], :], x[:, s[1], :],
             self.bucket_embedding(abs(s[1] - s[0] + 1), device).repeat(
                 (batch_size, 1)).to(device)),
            dim=1) for s in span_indices]

        return torch.stack(spans, dim=1), span_indices


class PrunedTargetOpinion:
    def __init__(self):
        pass

    def __call__(self, spans_probability, nz):
        target_indices = torch.topk(spans_probability[:, :, SpanLabel.ASPECT.value], nz, dim=-1).indices
        opinion_indices = torch.topk(spans_probability[:, :, SpanLabel.OPINION.value], nz, dim=-1).indices
        return target_indices, opinion_indices


class TargetOpinionPairRepresentation(nn.Module):
    def __init__(self, distance_embeddings_dim):
        super(TargetOpinionPairRepresentation, self).__init__()
        self.bucket_bins = [0, 1, 2, 3, 4, 5, 7, 8, 15, 16, 31, 32, 63, 64]
        self.distance_embeddings = nn.Embedding(len(self.bucket_bins), distance_embeddings_dim)

    def min_distance(self, a, b, c, d):
        return min(abs(b - c), abs(a - d))

    def bucket_embedding(self, width, device):
        em = [ix for ix, v in enumerate(self.bucket_bins) if width >= v][-1]
        return self.distance_embeddings(torch.LongTensor([em]).to(device))

    def forward(self, spans, span_indices, target_indices, opinion_indices):
        batch_size = spans.size(0)
        device = spans.device

        # candidate_indices :[(a,b,c,d)]
        # relation_indices :[(span1,span2)]
        candidate_indices, relation_indices = [], []
        for batch in range(batch_size):
            pairs = list(itertools.product(target_indices[batch].cpu().tolist(), opinion_indices[batch].cpu().tolist()))
            relation_indices.append(pairs)
            candidate_ind = []
            for pair in pairs:
                a, b = span_indices[pair[0]]
                c, d = span_indices[pair[1]]
                candidate_ind.append((a, b, c, d))
            candidate_indices.append(candidate_ind)

        candidate_pool = []
        for batch in range(batch_size):
            # Handle the case when relation_indices[batch] is empty
            if len(relation_indices[batch]) == 0:
                # Create an empty tensor with the expected shape
                # Assuming this is a valid shape based on your model
                candidate_pool.append(torch.zeros((0, spans.size(-1) * 2 + self.distance_embeddings.embedding_dim), device=device))
                continue
                
            relations = [
                torch.cat((spans[batch, c[0], :], spans[batch, c[1], :],
                           self.bucket_embedding(
                               self.min_distance(*span_indices[c[0]], *span_indices[c[1]]), device).squeeze(0))
                          , dim=0) for c in
                relation_indices[batch]]
            candidate_pool.append(torch.stack(relations))

        if all(p.size(0) == 0 for p in candidate_pool):
            # If all batches have empty candidate pools, return an empty tensor
            return torch.zeros((batch_size, 0, spans.size(-1) * 2 + self.distance_embeddings.embedding_dim), device=device), candidate_indices, relation_indices
            
        return torch.stack(candidate_pool), candidate_indices, relation_indices


class SpanAsteModel(nn.Module):
    def __init__(
            self,
            pretrain_model,
            target_dim,
            relation_dim,
            ffnn_hidden_dim=150,
            span_width_embedding_dim=20,
            span_maximum_length=8,
            span_pruned_threshold=0.5,
            pair_distance_embeddings_dim=128,
            device="cpu"
    ):
        super(SpanAsteModel, self).__init__()
        self.span_pruned_threshold = span_pruned_threshold
        self.pretrain_model = pretrain_model
        self.device = device

        # Use appropriate model type based on pretrain_model
        if "indobert-lite" in pretrain_model:
            print(f"Using AlbertModel for {pretrain_model}")
            self.bert = AlbertModel.from_pretrained(pretrain_model)
        else:
            print(f"Using BertModel for {pretrain_model}")
            self.bert = BertModel.from_pretrained(pretrain_model)
            
        encoding_dim = self.bert.config.hidden_size

        self.span_representation = SpanRepresentation(span_width_embedding_dim, span_maximum_length)
        span_dim = encoding_dim * 2 + span_width_embedding_dim
        self.span_ffnn = torch.nn.Sequential(
            nn.Linear(span_dim, ffnn_hidden_dim, bias=True),
            nn.ReLU(),
            nn.Dropout(p=0.4),
            nn.Linear(ffnn_hidden_dim, ffnn_hidden_dim, bias=True),
            nn.ReLU(),
            nn.Dropout(p=0.4),
            nn.Linear(ffnn_hidden_dim, target_dim, bias=True),
            nn.Softmax(-1)
        )
        self.pruned_target_opinion = PrunedTargetOpinion()
        self.target_opinion_pair_representation = TargetOpinionPairRepresentation(pair_distance_embeddings_dim)
        pairs_dim = 2 * span_dim + pair_distance_embeddings_dim
        self.pairs_ffnn = torch.nn.Sequential(
            nn.Linear(pairs_dim, ffnn_hidden_dim, bias=True),
            nn.ReLU(),
            nn.Dropout(p=0.4),
            nn.Linear(ffnn_hidden_dim, ffnn_hidden_dim, bias=True),
            nn.ReLU(),
            nn.Dropout(p=0.4),
            nn.Linear(ffnn_hidden_dim, relation_dim, bias=True),
            nn.Softmax(-1)
        )
        self.reset_parameters()

    def reset_parameters(self):
        for name, param in self.span_ffnn.named_parameters():
            if "weight" in name:
                nn.init.xavier_normal_(param)
        for name, param in self.pairs_ffnn.named_parameters():
            if "weight" in name:
                nn.init.xavier_normal_(param)

    def forward(self, input_ids, attention_mask, token_type_ids, seq_len):
        batch_size, sequence_len = input_ids.size()
        batch_max_seq_len = max(seq_len)

        # Disable dropout during inference
        self.span_ffnn.eval()
        self.pairs_ffnn.eval()

        bert_output = self.bert(input_ids, attention_mask, token_type_ids)
        x = bert_output.last_hidden_state
        spans, span_indices = self.span_representation(x, batch_max_seq_len)
        spans_probability = self.span_ffnn(spans)
        nz = int(batch_max_seq_len * self.span_pruned_threshold)

        target_indices, opinion_indices = self.pruned_target_opinion(spans_probability, nz)

        # Handle case where no candidates are found
        candidates, candidate_indices, relation_indices = self.target_opinion_pair_representation(
            spans, span_indices, target_indices, opinion_indices)

        # Handle case with no candidates
        if candidates.size(1) == 0:
            # Return empty tensors with expected shape
            candidate_probability = torch.zeros((batch_size, 0, len(RelationLabel)), device=self.device)
        else:
            candidate_probability = self.pairs_ffnn(candidates)

        # batch span indices
        span_indices = [span_indices for _ in range(batch_size)]

        return spans_probability, span_indices, candidate_probability, candidate_indices

# Load model and tokenizer
def load_model_and_tokenizer():
    print(f"Loading model from {MODEL_PATH}")
    print(f"Using {DEVICE} device")
    
    # Load tokenizer
    print(f"Loading tokenizer from {BERT_MODEL}")
    try:
        tokenizer = AutoTokenizer.from_pretrained(BERT_MODEL)
        print(f"Successfully loaded tokenizer: {type(tokenizer).__name__}")
    except:
        tokenizer = BertTokenizer.from_pretrained(BERT_MODEL)
        print(f"Fallback to BertTokenizer: {type(tokenizer).__name__}")
    
    # Define model structure
    target_dim, relation_dim = len(SpanLabel), len(RelationLabel)
    
    # Initialize model
    model = SpanAsteModel(
        BERT_MODEL,
        target_dim,
        relation_dim,
        device=DEVICE
    )
    
    # Load model state
    model_file_path = os.path.join(MODEL_PATH, "model.pt")
    if not os.path.exists(model_file_path):
        raise FileNotFoundError(f"Model file not found at {model_file_path}")
    
    # Load state dictionary
    model.load_state_dict(torch.load(model_file_path, map_location=torch.device(DEVICE)))
    model.to(DEVICE)
    model.eval()
    
    return model, tokenizer

# Function to predict triplets from a sentence
def predict_triplets(model, tokenizer, text):
    model.eval()
    with torch.no_grad():
        # Tokenize input
        tokens = ["[CLS]"] + tokenizer.tokenize(text) + ["[SEP]"]
        print(f"\nTokens: {tokens}")
        
        # Convert to model inputs
        inputs = tokenizer(text, max_length=128, padding=True, truncation=True, return_tensors="pt")
        input_ids = inputs.input_ids.to(DEVICE)
        attention_mask = inputs.attention_mask.to(DEVICE)
        token_type_ids = inputs.token_type_ids.to(DEVICE)
        seq_len = (input_ids != 0).sum().item()
        
        try:
            # Forward pass
            spans_probability, span_indices, relations_probability, candidate_indices = model(
                input_ids, attention_mask, token_type_ids, [seq_len])
            
            # Extract triplets
            relations_probability = relations_probability.squeeze(0)
            predict = []
            seen_pairs = set()
            
            for idx, can in enumerate(candidate_indices[0]):
                a, b, c, d = can
                
                # Skip if confidence is too low
                if relations_probability[idx].max().item() < 0.3:  # Confidence threshold
                    continue
                    
                # Extract and clean spans
                aspect = tokenizer.convert_tokens_to_string(tokens[a:b+1])
                opinion = tokenizer.convert_tokens_to_string(tokens[c:d+1])
                
                # Clean special tokens and extra chars
                aspect = aspect.replace('[CLS]', '').replace('[SEP]', '').strip()
                opinion = opinion.replace('[CLS]', '').replace('[SEP]', '').strip()
                
                # Remove trailing punctuation
                aspect = aspect.rstrip('.,!?;:')
                opinion = opinion.rstrip('.,!?;:')
                
                # Apply specialized cleaning
                aspect = clean_aspect_term(aspect)
                opinion = clean_opinion_term(opinion)
                
                # Skip empty spans
                if not aspect or not opinion:
                    continue
                    
                # Get sentiment
                sentiment = RelationLabel(relations_probability[idx].argmax(-1).item()).name
                confidence = relations_probability[idx].max().item()
                
                # Debug output
                print(f"Candidate: Aspect: '{aspect}', Opinion: '{opinion}', Sentiment: {sentiment}, Confidence: {confidence:.4f}")
                
                if sentiment != RelationLabel.INVALID.name:
                    pair = (aspect, opinion, sentiment)
                    if pair not in seen_pairs:
                        predict.append((aspect, opinion, sentiment, confidence))
                        seen_pairs.add(pair)
            
            # Create a DataFrame for better display
            if predict:
                df = pd.DataFrame(predict, columns=["Aspect", "Opinion", "Sentiment", "Confidence"])
                df = df.sort_values(by="Confidence", ascending=False)
                # Drop duplicates keeping first (highest confidence)
                df = df.drop_duplicates(subset=["Aspect", "Opinion"], keep="first")
                
                # Apply additional cleanup for better results
                final_df = pd.DataFrame(columns=["Aspect", "Opinion", "Sentiment", "Confidence"])
                seen_aspect_pairs = set()
                
                for _, row in df.iterrows():
                    # Force first word only for common aspects
                    aspect = row["Aspect"]
                    words = aspect.split()
                    
                    if len(words) > 1 and words[0].lower() in ['aplikasi', 'sistem', 'fitur', 'layanan', 'app']:
                        aspect = words[0]
                        
                    # Only keep first occurrence of each aspect
                    aspect_pair = (aspect, row["Opinion"])
                    if aspect_pair not in seen_aspect_pairs:
                        final_df = pd.concat([final_df, pd.DataFrame({
                            "Aspect": [aspect],
                            "Opinion": [row["Opinion"]],
                            "Sentiment": [row["Sentiment"]],
                            "Confidence": [row["Confidence"]]
                        })], ignore_index=True)
                        seen_aspect_pairs.add(aspect_pair)
                        
                if not final_df.empty:
                    return final_df
                    
            # Default empty DataFrame if no predictions
            return pd.DataFrame(columns=["Aspect", "Opinion", "Sentiment", "Confidence"])
            
        except Exception as e:
            print(f"Error during prediction: {e}")
            return pd.DataFrame(columns=["Aspect", "Opinion", "Sentiment", "Confidence"])
```

```python

# Main execution section
try:
    model, tokenizer = load_model_and_tokenizer()
    print("Model and tokenizer loaded successfully!")
    
    # Function for interactive predictions
    def predict_custom_text(text):
        print(f"\nPredicting triplets for: '{text}'")
        result = predict_triplets(model, tokenizer, text)
        print("\nFinal Prediction Results:")
        return result
    
    # Example usage
    example_text = "aplikasi nya bagus, tapi proses login lama sekali."
    result = predict_custom_text(example_text)
    display(result)
    
    # Now you can use predict_custom_text() for any custom text
    print("\nYou can now predict triplets for any custom text using predict_custom_text()")
    
except Exception as e:
    print(f"Error during model loading or prediction: {e}")
    # Print more detailed stack trace
    import traceback
    traceback.print_exc()
```
